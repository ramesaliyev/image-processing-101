# Histogram Creation and Equalization

**Initialization**

```python
import os
import sys

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

import lib.pgm as libpgm
```

**Read PGM**

```python
pgm_path = '../assets/pgms/lena.ascii.pgm'
pgm = libpgm.read(pgm_path)
libpgm.show(pgm)
```

**Calculate and Plot Histogram**

```python
_, intensity = libpgm.calculate_histogram(pgm);
```

```python
import matplotlib.pyplot as plt

plt.hist(range(256), 256, weights=intensity)
plt.show()
```

**Do Histogram Equalization**

```python
plt.rcParams['figure.figsize'] = [10, 5]
plt.subplot(1, 2, 1)
libpgm.show(pgm)        

libpgm.equalize_histogram(pgm)

plt.subplot(1, 2, 2)
libpgm.show(pgm)
```

**Recalculate Histogram**

```python
_, newintensity = libpgm.calculate_histogram(pgm);
plt.subplot(1, 2, 1)
plt.hist(range(256), 256, weights=intensity)
plt.subplot(1, 2, 2)
plt.hist(range(256), 256, weights=newintensity)
plt.show()
```
