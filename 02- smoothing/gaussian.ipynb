# Gaussian Filter

**Initialization**

```python
import os
import sys

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

import math
import numpy as np
import matplotlib.pyplot as plt
    
import lib.pgm as libpgm
import lib.conv as libconv
import lib.smooth as libsmooth
import lib.norm as libnorm
import lib.helpers as libhelpers
```

**Read PGM**

```python
pgm_path = '../assets/pgms/lena.ascii.pgm';
pgm = libpgm.read(pgm_path)
```

**Helper Functions**

```python
def apply_gaussian_filter(image, kernel_size, kernel_sigma):
    kernel = libsmooth.create_gaussian_kernel(kernel_size, kernel_sigma)
    return libconv.convolve2d(image, kernel);
```

## Apply Gaussian Filter

```python
plt.figure(figsize=(15, 15), dpi=72)
plt.subplot(2, 2, 1)
plt.gca().set_title('Original')
libpgm.show(pgm)  

# 1
kernel_size = 5
kernel_sigma = 5.5
image = libhelpers.pgm_to_image(pgm)
image = apply_gaussian_filter(image, kernel_size, kernel_sigma)
image = libnorm.apply_minmax_nd(image)

plt.subplot(2, 2, 2)
plt.gca().set_title('Gaussian Filter' + ' Size=' + str(kernel_size) + ' Sigma=' + str(kernel_sigma))
libhelpers.show_image(image)

# 2
kernel_size = 7
kernel_sigma = 5
image = libhelpers.pgm_to_image(pgm)
image = apply_gaussian_filter(image, kernel_size, kernel_sigma)
image = libnorm.apply_minmax_nd(image)

plt.subplot(2, 2, 3)
plt.gca().set_title('Gaussian Filter' + ' Size=' + str(kernel_size) + ' Sigma=' + str(kernel_sigma))
libhelpers.show_image(image)

# 3
kernel_size = 7
kernel_sigma = 7
image = libhelpers.pgm_to_image(pgm)
image = apply_gaussian_filter(image, kernel_size, kernel_sigma)
image = libnorm.apply_minmax_nd(image)

plt.subplot(2, 2, 4)
plt.gca().set_title('Gaussian Filter' + ' Size=' + str(kernel_size) + ' Sigma=' + str(kernel_sigma))
libhelpers.show_image(image)
```
