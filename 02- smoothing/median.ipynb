# Median Filter

**Initialization**

```python
import os
import sys

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

import math
import numpy as np
import matplotlib.pyplot as plt
    
import lib.pgm as libpgm
import lib.conv as libconv
import lib.point as libpoint
import lib.smooth as libsmooth
```

**Read PGM**

```python
pgm_path = '../assets/pgms/lena2.ascii.pgm';
pgm = libpgm.read(pgm_path)
```

**Helper Functions**

```python
def pgm_to_image(pgm):
    (pixels, dims, *_) = pgm[:]
    return np.array(pixels).reshape(dims).tolist()

def show_image(image):
    plt.imshow(image, cmap='gray', vmin=0, vmax=255)
```

## Apply Median Filter

```python
plt.figure(figsize=(15, 15), dpi=72)
plt.subplot(2, 2, 1)
plt.gca().set_title('Original')
libpgm.show(pgm)  

# 1
kernel_size = 3
image = pgm_to_image(pgm)
image = libsmooth.apply_median_filter(image, kernel_size)
map(libpoint.apply_minmax, image)

plt.subplot(2, 2, 2)
plt.gca().set_title('Median Filter' + ' Size=' + str(kernel_size))
show_image(image)

# 2
kernel_size = 5
image = pgm_to_image(pgm)
image = libsmooth.apply_median_filter(image, kernel_size)
map(libpoint.apply_minmax, image)

plt.subplot(2, 2, 3)
plt.gca().set_title('Median Filter' + ' Size=' + str(kernel_size))
show_image(image)

# 3
kernel_size = 7
image = pgm_to_image(pgm)
image = libsmooth.apply_median_filter(image, kernel_size)
map(libpoint.apply_minmax, image)

plt.subplot(2, 2, 4)
plt.gca().set_title('Median Filter' + ' Size=' + str(kernel_size))
show_image(image)
```
